```python
!pip install pandas
import pandas as pd

# Load  dataset
df = pd.read_csv('books_dataset.csv')
df.head(3)
```

# Data cleaning and preprocessing


```python
print(df.isnull().sum())
df.dropna(inplace=True)  
df.drop_duplicates(inplace=True)  

```

```python
df_hp = df[df['original_title'].str.contains('Harry Potter', case=False, na=False)]

print("Unique Harry Potter Books:")
print(df_hp['original_title'].unique())
```

```python
# Focus on Harry Potter series
harry_potter_df = df[df['title'].str.contains('Harry Potter')]
harry_potter_df
```

# Most Selling Books within the Harry Potter Series

```python
most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False)
most_selling_books.head(5)
```

# Calculate the average rating of the Harry Potter books

```python
average_rating = harry_potter_df['average_rating'].mean()
average_rating
```
